# Polymorphism in user defined classes
Method Overriding

```python
class India:

    def capital(self):
        print("New Delhi is capital of India")

    def language(self):
        print("Hindi is widely spoken in India")
```

```python
class USA:

    def capital(self):
        print("Washington D.C. is capital of USA")

    def language(self):
        print("English is widely spoken in USA")
```

```python
class France:

    def capital(self):
        print("Paris is capital of France")

    def language(self):
        print("French is widely spoken in France")
```

```python
c1 = India()
type(c1)
```

```python
c1.capital()
```

```python
c1.language()
```

```python
c2 = USA()
type(c2)
```

```python
c2.capital()
```

```python
c2.language()
```

Easily apply for loop on multiple classes to get results

```python
c = [India(), USA(), France()]
c
```

```python
for i in c:
    print(type(i))
    i.language()
    i.capital()
    print("\n","="*50, "\n")
```

Calulate area and perimeter for different shapes

```python
from dataclasses import dataclass
```

```python
@dataclass
class Rectangle:
    height: int | float
    width: int | float

    def perimeter(self):
        return 2 * (self.height + self.width)
    
    def area(self):
        return self.width * self.height
```

```python
from math import pi
```

```python
pi
```

```python
from dataclasses import dataclass
```

```python
@dataclass
class Circle:
    radius: int | float

    def perimeter(self):
        return 2 * pi * self.radius
    
    def area(self):
        return pi * (self.radius ** 2)
```

```python
shapes = [
    Triangle(3, 4, 5),
    Circle(7),
    Rectangle(10, 20),
    Circle(20),
    Triangle(12, 13, 14),
    Rectangle(15, 12)
]
```

```python
shapes
```

```python
for i in shapes:
    print(i, type(i))
    print(f"Perimeter : {i.perimeter():.2f}")
    print(f"Area : {i.area():.2f}")
    print("\n", "="*50, "\n")
```
