```python
def division():
    num = float(input("Please enter Numerator : "))
    den = float(input("Please enter Denominator : "))
    return num / den 
```

```python
# num : 3, den : 4
print(division())
print("Program Complete")
```

```python
# num: 5, den: 0
print(division())
print("Program Complete")
```

```python
# Num - 43, Den - abc
print(division())
print("Program Complete")
```

```python
# Num - Blank
print(division())
print("Program Complete")
```

Generic way of exception handling

```python
def division2():
    try:
        num = float(input("Please enter numerator : "))
        den = float(input("Please enter denominator : "))
        return num / den
    except Exception as e:
        print(f"Exception Occured : {e}")
```

```python
# num - 10, den - 7
print(division2())
print("Program Complete")
```

```python
# num - 11, den - 0
print(division2())
print("Program Complete")
```

```python
# num - abc
print(division2())
print("Program Completed")
```

```python
# num - 123, den - blank
print(division2())
print("Program Complete")
```

Method 2 - Handling exceptions for each erros

```python
def division3():
    try:
        num = float(input("Please enter numerator : "))
        den = float(input("Please enter denominator : "))
        return num / den
    except ZeroDivisionError:
        print("Denominator cannot be 0")
    except ValueError:
        print("Please enter a valid number strings and blank values are not allowed")   
```

```python
# num - 7, den - 8
print(division3())
print("Program Completed")
```

```python
# num - 23, den - 0
print(division3())
print("Program Completed")
```

```python
# num - 12, den - b
print(division3())
print("Program Completed")
```

```python
# num - blank
print(division3())
print("Program Completed")
```
